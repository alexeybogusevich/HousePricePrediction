```python
#for each of the categorical features above we will calculate ANOVA to programmatically 
#define if this feature is significant for sale price 
```

```python
cat_features = categorical_data.columns
df_train[cat_features] = df_train[cat_features].fillna('Missing') #else ANOVA wont run

#Onward
anova = {'feature':[], 'f':[], 'p': []}
for cat in cat_features:
    group_prices = []
    for group in df_train[cat].unique():
        group_prices.append(df_train[df_train[cat] == group]['SalePrice'].values)
    f, p = scipy.stats.f_oneway(*group_prices)
    anova['feature'].append(cat)
    anova['f'].append(f)
    anova['p'].append(p)
    
anova = pd.DataFrame(anova)
anova = anova[['feature','f','p']]
anova.sort_values('p', inplace=True)

# p < 0.05 => significant
# if we get even single [p < 0.05] then we reject N0 hypothesis, which is:
# Our N0 = categorical features are all tent to have almost same significance for saleprice definition
```

```python
anova # p < 0.05
```

```python
# from above only three last (street, MoSold, YrSold) are not significant
```

```python
# Neighborhood, Foundation, GarageType, MSSubClass are very significant, they all define sale price
```

```python
plt.figure(figsize=(8,5))
sns.barplot(anova.feature, anova['p'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(anova.feature, np.log(1./anova['p']))
plt.xticks(rotation=90)
plt.show() #significance >>>>>
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(model_data.copy(), y, test_size=0.3, random_state=42)
print('Shapes: ')
print('X_train: ', X_train.shape)
print('X_test: ', X_test.shape)
print('y_train: ', y_train.shape)
print('y_test: ', y_test.shape)
```

```python
#1019 records for training
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler
```

```python
stdsc = StandardScaler()

num_features.remove('Id')
num_features.remove('SalePrice')

# numerical features standartization - their means will become zero with the standard deviation of 1
# this will help weights to efficiently learn

X_train.loc[:, num_features] = stdsc.fit_transform(X_train[num_features])
X_test.loc[:, num_features] = stdsc.transform(X_test[num_features])
```

```python
from sklearn.model_selection import cross_val_score

def rsme(model, X, y):
    cv_scores = -cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=10)
    return np.sqrt(cv_scores)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1., 5., 10., 25.], 'max_iter':[50000]}
lasso = GridSearchCV(Lasso(), cv=5, param_grid=param_grid, scoring='neg_mean_squared_error') #cv - cross-validation parameter
#each time the set is divided into 5 parts and validated to search for optimal parameters
lasso.fit(X_train, y_train)
#Lasso regresion at first will help us to drop insignificant columns by finding optimal parameters
alpha = lasso.best_params_['alpha']

param_grid = {'alpha':[x/100. * alpha for x in range(50, 150, 5)], 'max_iter':[50000]}
lasso = GridSearchCV(Lasso(), cv=5, param_grid=param_grid, scoring='neg_mean_squared_error')
lasso.fit(X_train, y_train)
alpha = lasso.best_params_['alpha']
lasso = lasso.best_estimator_

lasso_train_rsme = rsme(lasso, X_train, y_train).mean()
lasso_test_rsme = rsme(lasso, X_test, y_test).mean()

print('Lasso --> Train RSME: {:.5f} | Test RSME: {:.5f} | alpha: {:.5f}'.format(lasso_train_rsme, lasso_test_rsme, alpha))
```

```python
# if train and test error are almost equal, then the conclusion is that the model just memorized answers,
# it did not understand the pattern
# overfitting
```

```python
coefs = pd.DataFrame({'coefs': lasso.coef_, 'Positive': lasso.coef_ > 0}, index=X_train.columns)
coefs['coefs_abs'] = np.abs(coefs.coefs)
print('Lasso dropped {} of {} features.'.format(sum(coefs.coefs == 0), coefs.shape[0]))

top_coefs = coefs.sort_values('coefs_abs', ascending=False).head(20)
plt.figure(figsize=(8,10))
sns.barplot(top_coefs.coefs_abs, top_coefs.index, orient='h', hue=top_coefs.Positive)
plt.title('LASSO Regression: Top Features')
plt.xlabel('Absolute coefficient')
plt.show()
```

```python
import xgboost as xgb
```

```python
regr = xgb.XGBRegressor(colsample_bytree=0.2,
                       gamma=0.0,
                       learning_rate=0.05,
                       max_depth=6,
                       min_child_weight=1.5,
                       n_estimators=7200,
                       reg_alpha=0.9,
                       subsample=0.2,
                       seed=42,
                       silent=1)

regr.fit(X_train[top_coefs.index], y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(y_test, y_pred))

y_pred = regr.predict(X_test[top_coefs.index])

print("XGBoost score on training set: ", rmse(y_test, y_pred))
```
